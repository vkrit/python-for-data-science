## Dependencies between Arrival Delay and Departure Delay

### Finding dependency between variables

We will find depen.....

```python
flights[['dep_delay','arr_delay']].corr()
```

```python
#Using seaborn package explore the dependency of arr_delay on dep_delay (scatterplot or regplot) for flights dataset
sns.jointplot(x='dep_delay', y='arr_delay', data=flights)
```

```python
#Using seaborn package explore the dependency of arr_delay on dep_delay (scatterplot or regplot) for flights dataset
sns.regplot(x='dep_delay', y='arr_delay', data=flights)
```
