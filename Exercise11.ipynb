```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
#Read csv file
flights = pd.read_csv("flights.csv")
```

```python
flights.columns
```

```python
flights = flights.dropna()
#fligths = flights.sample(n=200, random_state=10 , replace=True)
```

```python
lm = smf.ols(formula='dep_delay ~ arr_delay', data=flights).fit()
mdl = lm.get_robustcov_results(cov_type='HAC',maxlags=1)
print(mdl.summary())
```

```python
print(lm.params)
print('R2: ', lm.rsquared)
```

```python
lm = smf.ols(formula='dep_delay ~ arr_delay + carrier', data=flights).fit()
mdl = lm.get_robustcov_results(cov_type='HAC',maxlags=1)
print(mdl.summary())
```

```python
flightsub = flights[['arr_delay','dep_delay','carrier']]
flightsubdum = pd.get_dummies(flightsub, prefix=['carrier'])
flightsubdum
```

```python
print('R2: ', lm.rsquared)
```

```python
lm = smf.ols(formula='dep_delay ~ arr_delay + carrier_AA + carrier_AS + carrier_DL + carrier_UA + carrier_US', data=flightsubdum).fit()
mdl = lm.get_robustcov_results(cov_type='HAC',maxlags=1)
print(mdl.summary())
```

```python
print('R2: ', lm.rsquared)
```
